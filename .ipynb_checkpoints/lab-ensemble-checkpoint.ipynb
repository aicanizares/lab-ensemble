```python
#Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship.head()
```

Now perform the same as before:
- Feature Scaling
- Feature Selection


```python
spaceship = spaceship.dropna()
```

```python
def extract_first_letter(value):
    return value[0]
```

```python
spaceship["Cabin"] = spaceship["Cabin"].apply(extract_first_letter)
```

```python
spaceship
```

```python
spaceship = spaceship.drop(["PassengerId", "Name"], axis=1)
```

```python
spaceship
```

```python
spaceship_dummies = pd.get_dummies(spaceship[["HomePlanet", "CryoSleep", "Cabin", "Destination"]])
```

```python
spaceship_dummies
```

```python
features_numerical = spaceship.select_dtypes(include=["number"])
```

```python
features_numerical
```

```python
features = pd.concat([features_numerical, spaceship_dummies], axis=1)
```

```python
features
```

```python
target = spaceship["Transported"]
```

```python
target
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.20, random_state=42)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
normalizer = MinMaxScaler()
```

```python
normalizer.fit(X_train)
```

```python
X_train_norm = normalizer.transform(X_train)

X_test_norm = normalizer.transform(X_test)
```

```python
X_train_norm
```

```python
X_test_norm = pd.DataFrame(X_test_norm, columns = X_test.columns)
X_test_norm.head()
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap=plt.cm.Blues)
plt.show()
```

```python
# True Negative: 546
# False positive: 107
# False negative: 189
# True Positive: 480
# Formulas: https://developers.google.com/machine-learning/crash-course/classification/accuracy-precision-recall?hl=es-419
```

```python
# Accuracy = Correct classifications (TP + TN) / Total classifications (TP + TN + FP + FN)
# Accuracy = (480 + 546) / Total classifications (480 + 546 + 107 + 189) = 0.77609682299
```

```python
from sklearn import metrics

actual = np.array([1] * 480 + [0] * 546 + [1] * 189 + [0] * 107)
predicted = np.array([1] * 480 + [0] * 546 + [0] * 189 + [1] * 107)
```

```python
accuracy = metrics.accuracy_score(actual, predicted)
```

```python
accuracy
```

```python
# Precision = Correctly classified actual positives (TP) / Everything classified as positive (TP + FP)
# Precision = 480 / (480 + 107) = 0.81771720613
```

```python
from sklearn.metrics import precision_score

actual = [1] * 480 + [0] * 107
predicted = [1] * (480 + 107)
```

```python
precision = precision_score(actual, predicted)

precision
```

```python
# Recall = Correctly classified actual positives (TP) / All actual positives (TP + FN) 
# Recall = 480 / (480 + 189) = 0.71748878923
```

```python
from sklearn.metrics import recall_score

actual = [1] * 480 + [1] * 189 + [0] * 107 + [0] * 546
predicted = [1] * 480 + [0] * 189 + [1] * 107 + [0] * 546
```

```python
recall = recall_score(actual, predicted)
```

```python
recall
```

```python
# F1 = False positive rate (FPR) = Incorretly classified actual negatives (FP) / All actual negatives (FP + TN)
# F1 = 107 / (107 + 546) = 0.16385911179
```

```python
actual = np.array([1] * 480 + [1] * 189 + [0] * 107 + [0] * 546)
predicted = np.array([1] * 480 + [0] * 189 + [1] * 107 + [0] * 546)
```

```python
FP = 107
TN = 546
FPR = FP / (FP + TN)
```
